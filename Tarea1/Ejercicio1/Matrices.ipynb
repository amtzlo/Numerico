```python
##Funciones

#1a)
def Multiplicacion_matriz(a,b):
  if len(a)==len(b[0]):# comparo el numero columnas de A contra el
                      # numero de filas de B
    c=[]
    for i in range(len(a)): # va sobre el numero de columnas de A
      col=[]
      for j in range(len(b[0])):# va sobre el numero de renglones de A
        MM=0.0 # MultiMatriz (MM) se inicia en cero
        for k in range(len(a[0])):
          MM= MM+ a[i][k]*b[k][j]
        col.append(MM)# cada entrada por columna de una fila
      c.append(col)# cada una de las filas
    return c
  else:
    print("No son de la misma dimensión")
    return None
#1a)
def multiplicacion_matrices():
    #Pedimos al usuario que nos de dos matrices
    boton_1 = input("propón una matriz de esta manera:[[a,b],[c,d],...] :")
    boton_2 = input("propón otra matriz de esta manera:[[a,b],[c,d],...] :")
    #Quiero convertir lo que el usuario ingrese en una matriz según python
    #La función eval() lo que hace es convertir una cadena en una lista de listas
    a = np.array(eval(boton_1))
    b = np.array(eval(boton_2))
    #Para verificar si el producto se puede hacer
    if a.shape[0] == b.shape[1]:
    # @: Está definido como la multiplicación matricial en numpy
        resultado = print("El resultado de la multiplicación matricial es:", a@b)
        return resultado
    else:
        print("Las matrices deben coincidir en el número de columnas y filas")

#1b)
def mult_x_bloq(M,B,row,Brow,col,Bcol):
    if M.shape[0] == B.shape[1]:
      # Para la matriz M
      M1=np.copy(M)
      M1=np.delete(M1,row,axis=0)# elimina sobre sobre los renglones
      M1=np.delete(M1,col,axis=1)# elimina sobre sobre los renglones
      # Para la matriz B
      B1=np.copy(B)
      B1=np.delete(B1,Brow,axis=0)# elimina sobre sobre los renglones
      B1=np.delete(B1,Bcol,axis=1)# elimina sobre sobre los renglones

      if M1.shape[0] == B1.shape[1]:
          return M1@B1
    else:
        print("No tienen el mismo número de filas y columnas")
#1c)
def Transpuesta():
    boton = input("Ingresa una matriz de la forma:[[a,b],[c,d],...]:")
    #eval() nos ayudará a convertir lo que meta el usuario en una lista
    mat = np.array(eval(boton))
    #.T está dentro de la librería de numpy y nos da la transpuesta de una matriz
    transpuesta = mat.T
    resultado = print("La matriz transpuesta es:",transpuesta)
    return resultado

#1d)
def potencia_mat():
    #Vamos a pedirle al usurio que ingrese una matriz y que elija
    #la potencia a la cual la quiere elevar (ojalá que no se manche)
    boton = input("Ingresa una matriz de la forma: A= [[a,b],[c,d],...]:")
    potencia = int(input("Elige la potencia a la cual la quieres elevar:"))
    matt = np.array(eval(boton))
    #Comprobar que la matriz tenga el mismo número de filas y columnas
    if matt.shape[0] == matt.shape[1]:
        if potencia == 0:
           # Si la potencia es 0 sabemos que nos tiene que devolver la identidad
           # Existe la función np.eye() para devolver una matriz identidad
           # Se usa matt.shape[0] dentro de la función identidad para que la matriz
           # identidad sea de la misma dimensión de la matriz matt
           resultado = np.eye(matt.shape[0])
           print(resultado)
        else:
           # np.linalg.matrix_power(matriz,potencia)Eleva la matriz a la potencia solicitada
           resultado = np.linalg.matrix_power(matt, potencia)
           return resultado
    else:
        print("La matriz tiene que ser cuadrada")

#1e)
def ProdVec(a,b):
  if len(a)==len(b):
      v1 = np.array(a)
      v2 = np.array(b)
      producto = np.cross(v1,v2)
      return producto
  else:
      print("Ambos vectores deben de ser tridimensionales")

#1f)
def rotacion_vect(a,angulo):
    #Definimos la variable en donde almacenaremos en angulo
    t = np.radians(angulo)
    #Definimos la matriz de rotación
    m_rot =  np.array([[np.cos(t), -np.sin(t)],
                       [np.sin(t), np.cos(t)]])
    #Toca hacer el producto de vector por matriz
    #Revisamos número de columnas y número de entradas
    if len(m_rot[0])==len(a):
        v_rot=[]
        for i in range(len(m_rot)):
            mv = 0.0
            #Cada fila será multiplicada por las entradas del vector y
            #se almacenarán en una lista
            for j in range(len(m_rot[0])):
                mv= mv+ m_rot[i][j]*a[j]
            v_rot.append(mv)
        return v_rot
    else:
        print("No se puede multiplicar")

#1g)
def rot_tri(v,angulo):
    #Definimos la variable en donde almacenaremos en angulo
    t = np.radians(angulo)
    #Definimos la matrices de rotación sobre cada eje x,y,z
    m_rot_x =  np.array([[1,0,0],
                         [0,np.cos(t), -np.sin(t)],
                         [0,np.sin(t), np.cos(t)]])
    m_rot_y = np.array([[np.cos(t),0,np.sin(t)],
                        [0,1,0],
                        [-np.sin(t),0,np.cos(t)]])
    m_rot_z = np.array([[np.cos(t),-np.sin(t),0],
                        [np.sin(t),np.cos(t),0],
                        [0,0,1]])
    #Toca hacer el producto de vector por cada matrizmatriz


    #Revisamos número de columnas y número de entradas
    if len(m_rot_x[0])==len(v):
        v_rot_x=[]
        for i in range(len(m_rot_x)):
            mvx = 0.0
            #Cada fila será multiplicada por las entradas del vector y
            #se almacenarán en una lista
            for j in range(len(m_rot_x[0])):
                mvx= mvx+ m_rot_x[i][j]*a[j]
            v_rot_x.append(mvx)
        return v_rot_x
    if len(m_rot_y[0])==len(v):
        v_rot_y=[]
        for i in range(len(m_rot_y)):
            mvy = 0.0
            #Cada fila será multiplicada por las entradas del vector y
            #se almacenarán en una lista
            for j in range(len(m_rot_y[0])):
                mvy= mvy+ m_rot_y[i][j]*a[j]
            v_rot_y.append(mvy)
        return v_rot_y
    if len(m_rot_z[0])==len(v):
        v_rot_z=[]
        for i in range(len(m_rot_z)):
            mvz = 0.0
            #Cada fila será multiplicada por las entradas del vector y
            #se almacenarán en una lista
            for j in range(len(m_rot_z[0])):
                mvz= mvz+ m_rot_z[i][j]*a[j]
            v_rot_z.append(mvz)
        return v_rot_z
    else:
        print("No se puede multiplicar")

#1h)
def norma_1(matriz):
    # Obtenemos las filas y colmumnas de la matriz
    filas = len(matriz)
    columnas = len(matriz[0])
    # Creamos una lista donde guardaremos la suma de columnas
    suma_cols = [0] * columnas

    # Iteramos sobre la matriz para calcular la suma de columns
    for j in range(columnas):
        for i in range(filas):
            suma_cols[j] += abs(matriz[i][j])

    # Regresamos el maximo de la suma de las columnas
    return max(suma_cols)

#1i)
def norma_inf(matriz):
    # Lista con la suma de cada fila de la matriz
    suma_filas = [sum(abs(x) for x in fila) for fila in matriz]

    # Regresamos el maximos de las sumas
    return max(suma_filas)

#1j)
def norma_2(matriz):
    # Calculamos el cuadrado de cada elemento en la matriz y sumamos
    cuadrados = sum(x**2 for fila in matriz for x in fila)

    # Regresamos la raiz cuadrada
    return math.sqrt(cuadrados)

```
