## Ejercicio 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Puntos dados en la partición
x_int  = np.array([0, np.pi/4, np.pi/2, np.pi, 5*np.pi/4])
y_int = np.cos(x_int)

# Generar puntos para graficar
x_cos = np.linspace(0, 5*np.pi/4, 500)
y_cos = np.cos(x_cos)  # Valores reales de la función cos(x)
```

```python
plt.figure(figsize=(10, 6))
plt.title('Interpolación lineal de f(x) = cos(x)', fontsize=14)
plt.plot(x_int, y_int, color='red', linestyle = '--', label = 'Interpolacion Lineal')
plt.scatter(x_int, y_int, color='red', label='Puntos de interpolación', zorder=5)
plt.plot(x_cos, y_cos, label = 'f(x) = cos(x)')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## Ejercicio 2

```python
x2_int = np.array([0, 1, 2,  3])
f_int = lambda x: (0.8455*x**3) - (1.0604 * x ** 2) + (1.9331 ** x) + 1.0
y2_int = np.array(list(map(f_int, x2_int)))

x2_f = np.linspace(0, 3, 500)
y2_f = np.array(list(map(f_int, x2_f)))

x2_exp = np.linspace(0, 3, 500)
y2_exp = np.exp(x2_exp)
```

```python
plt.figure(figsize=(10, 6))
plt.title('Interpolación de f(x) = exp(x)', fontsize=14)
plt.plot(x2_f, y2_f, color='red', linestyle = '--', label = 'Interpolacion de Langrage')
plt.scatter(x2_int, y2_int, color='red', label='Puntos de interpolación', zorder=5)
plt.plot(x2_exp, y2_exp, label = 'f(x) =exp(x))')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
